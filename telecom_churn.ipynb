H√° uma pequena diferen√ßa no total de minutos gastos e no total cobrado entre o dia e noite. Os minutos totais pela noite √© maior que o do dia e o total cobrado pela noite √© menor do que √© cobrado durante o dia.   

```python
noite = treino.total_night_minutes.sum()
dia = treino.total_day_minutes.sum()
pizza = list([dia,noite])

labels = 'Total de minutos no periodo da manh√£','Total de minutos no periodo da noite'
explode = (0,0.1)

fig1,ax1 = plt.subplots()
ax1.pie(pizza,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal')
plt.show()
```

```python
noite = treino.total_night_charge.sum()
dia = treino.total_day_charge.sum()
pizza = list([dia,noite])

labels = 'Total cobrado por liga√ß√µes no periodo da manh√£','Total cobrado por liga√ß√µes no periodo da noite'
explode = (0,0.1)

fig1,ax1 = plt.subplots()
ax1.pie(pizza,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal')
plt.show()
```

```python
def limpadado(dataframe):
    treino = dataframe
    
    treino.international_plan.replace(('yes','no'),(1,0),inplace=True)
    treino.voice_mail_plan.replace(('yes','no'),(1,0),inplace=True)
    treino.churn.replace(('yes','no'),(1,0),inplace=True)
    return treino
```

```python
treino = limpadado(treino)
```

```python
treino['area_code'] = treino['area_code'].map(lambda x: x.lstrip('area_code_'))
```

```python
colunas = ['account_length','area_code','international_plan',                 
'voice_mail_plan',                    
'number_vmail_messages',              
'total_day_minutes',                
'total_day_calls',                    
'total_day_charge',                 
'total_eve_minutes',                
'total_eve_calls',                    
'total_eve_charge',                 
'total_night_minutes',              
'total_night_calls',                  
'total_night_charge',               
'total_intl_minutes',               
'total_intl_calls',                   
'total_intl_charge',                
'number_customer_service_calls']
dfteste = pd.DataFrame(novo_teste,columns=colunas)
```

```python
#como recomendado na instru√ß√£o desse projeto usei o LogisticRegression.
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

modelo = LogisticRegression()
modelo.fit(dftreino,y_treino)

previsoes = modelo.predict(dfteste)
matrix = confusion_matrix(y_test,previsoes)
report = classification_report(y_test,previsoes)


print(matrix)
print(report)
```

```python
#depois tentei melhora-lo com gridsearchcv como visto no cap06.
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')

valores_grid = {'solver': ['newton-cg', 'lbfgs', 'liblinear'],'penalty':['none','l1','l2','elasticnet'],'C':[0.001,0.01,0.1,1,10,100,1000]}

grid = GridSearchCV(estimator=modelo,param_grid=valores_grid)

grid.fit(dftreino,y_treino)

print('Acur√°cia: %.3f' % (grid.best_score_ * 100))
print('Melhores parametros do Modelo:\n',grid.best_estimator_)
```

```python
previsoes = grid.predict(dfteste)
```
